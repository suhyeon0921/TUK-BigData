```python
# 데이터 프레임(df)의 고객명, 성별, 나이만 10개의 데이터 출력
df[["고객명","성별","나이"]].head(10)
```

```python
# 1. 데이터프레임을 고객명순으로 정렬
# 2. 정렬한 데이터프레임을 고객명,성별, 나이만 10의 데이터 출력

sort_df = df.sort_values( by=["고객명"] )
sort_df[["고객명","성별","나이"]].head(10)
```

```python
# 1. 고객명, 나이순으로 정렬
# 2. 고객명, 직업, 나이항목만 20개 행을 출력

sort_df = df.sort_values(by=["고객명","나이"], ascending=[True, False] )
sort_df[["고객명","직업","나이"]].head(20)
```

```python
# 키가 165이상 고객들만 출력
# df[(df["키"]>=165)]
# 키가 165이상이고 몸무게가 50이상인 고객들만 출력

df[(df["키"] >= 165) & (df["몸무게"] >= 50)]
```

```python
# 고객명에 '김'이라는 문자가 들어가있는 고객 출력
df[df["고객명"].str.contains("김")]
```

```python
# 연령대 추가
df["연령대"] = (df["나이"] // 10) * 10

# 고객명, 성별, 나이, 연령대 출력
df[["고객명","성별","나이","연령대"]].head(10)
```

```python
# 연령대별로 고객수
result = df.groupby(by=["연령대","성별"]).size()
result
```

```python
# 연령대별 주사용ott별 고객수
result_ott = df.groupby(by=["연령대","주사용OTT"]).size()
result_ott
```
